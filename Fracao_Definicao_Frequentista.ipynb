# Fração de Caras em uma Moeda Pela Definição Frequentista Usando Python

```
import numpy as np
```

```
import matplotlib.pyplot as plt
```

```
#chances de sair cara
p = 0.6
```

```
#lista para armazenar a frequência de ocorrências
vp = []
```

```
#lista para armazenar o número de simulações
vsim = []
```

```
#número máximo de simulações
vmax = 1000
```

```
for nsim in np.arange(1, vmax, 10):
  nhead = 0 #número de caras
  for i in range(1, nsim): #o número de simulações aumenta
    if(np.random.uniform() < p):
      nhead += 1
  vp.append(nhead/nsim)
  vsim.append(nsim)
```

```
#mostrar os resultados em um gráfico
plt.figure(figsize=(8, 6))
plt.plot(vsim, vp, linestyle = '-', color = 'blue', label = 'Valor Simulado')
plt.axhline(y = p, color = 'r', linestyle = '--', label = 'Valor teórico')
plt.ylabel("Fração de caras", fontsize = 20)
plt.xlabel("Número de experimentos", fontsize = 20)
plt.xlim([0.0, vmax])
plt.ylim([0.0, 1.0])
plt.legend()
plt.show(True)
```
